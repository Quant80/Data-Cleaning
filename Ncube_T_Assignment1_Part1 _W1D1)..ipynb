### Conversion and Renaming Columns

```python
cols = data.columns
data[cols] = data[cols].apply(pd.to_numeric, errors='coerce')
```

```python
data.dtypes
```

```python
data.rename(columns={"Y":"Year","M":"Month","D":"Day","T":"Avg_Temp","TM":"Max_Temp","Tm":"Min_Temp","SLP":"Avg_AtP","STP":"Avg_HPA","H":"Humidity","PP":"Rainfaill","VV":"Visibility","V":"Wind_SPD","VM":"Max_Wind_SPD","VG":"Max_SPD_KMH","FG":"FOG","RA":"Rain","SN":"Snow","GR":"Hail","TS":"Thunder","TR":"Tornado"},inplace=True)
```

```python
data.head(2)
```

### Removing missing Values 

```python
data.dropna(subset=["Avg_Temp"], inplace=True)
data.shape
```

### Converting all date columns to a date data type

```python
pd.to_datetime(data[['Year', 'Month', 'Day']])
```
